```python
def plot_training(path, title):
    progress = pd.read_csv("experiment/" + path + "/progress.csv")
    progress.plot(title=title, logy = True,x='step', y=['grad_norm', "loss", "loss_q0", "loss_q1", "loss_q2", "loss_q3"])
```

```python
def plot_loss(path, title):
    progress = pd.read_csv("experiment/" + path + "/progress.csv")
    progress.plot(title=title, logy = True, y=["loss", "loss_q0", "loss_q1", "loss_q2", "loss_q3"])
```
